```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

## RR domain scanning

```python
CheB = pd.read_csv("../data/e_cloacae/hmm_profile/CheB_methylest_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
GerE =  pd.read_table("../data/e_cloacae/hmm_profile/GerE_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HTH_11 = pd.read_csv("../data/e_cloacae/hmm_profile/HTH_11_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HTH_18 = pd.read_csv("../data/e_cloacae/hmm_profile/HTH_18_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HTH_AraC = pd.read_csv("../data/e_cloacae/hmm_profile/HTH_AraC_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
LysR = pd.read_csv("../data/e_cloacae/hmm_profile/LysR_subs_out1", skiprows=5, header=None, skipfooter=9, sep=r"\s+", engine='python')
LytR = pd.read_csv("../data/e_cloacae/hmm_profile/LytTR_out1", skiprows=5,  header=None, skipfooter=9, sep=r"\s+", engine='python')
Response_reg = pd.read_csv("../data/e_cloacae/hmm_profile/Response_reg_out1", skiprows=5, skipfooter=9,header=None, sep=r"\s+", engine='python')
Trans_reg= pd.read_csv("../data/e_cloacae/hmm_profile/Trans_reg_C_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
CitT= pd.read_csv("../data/e_cloacae/hmm_profile/CitT_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
HTH_IcIR= pd.read_csv("../data/e_cloacae/hmm_profile/HTH_IclR_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
```

## HK domain scanning

```python
HAMP = pd.read_csv("../data/e_cloacae/hmm_profile/HAMP_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HATPase_c =  pd.read_table("../data/e_cloacae/hmm_profile/HATPase_c_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HATPase_c5 =  pd.read_table("../data/e_cloacae/hmm_profile/HATPase_c_5_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HisKA = pd.read_csv("../data/e_cloacae/hmm_profile/HisKA_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
His_kinase = pd.read_csv("../data/e_cloacae/hmm_profile/His_kinase_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
Hpt = pd.read_csv("../data/e_cloacae/hmm_profile/Hpt_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
PAS = pd.read_csv("../data/e_cloacae/hmm_profile/PAS_out1", skiprows=5, skipfooter=9, header=None, sep=r"\s+", engine='python')
HisKA_3= pd.read_csv("../data/e_cloacae/hmm_profile/HisKA_3_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
GAF = pd.read_csv("../data/e_cloacae/hmm_profile/GAF_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
MASE1 = pd.read_csv("../data/e_cloacae/hmm_profile/MASE1_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
CSK = pd.read_csv("../data/e_cloacae/hmm_profile/2CSK_N_out1", skiprows=5, skipfooter=9,  header=None, sep=r"\s+", engine='python')
```

## Cutoff by e-value

```python
HAMP.head()
```

```python
CheB_evalue= CheB[CheB[2]<0.01]
GerE_evalue=GerE[GerE[2]<0.01]
HTH_11_evalue=HTH_11[HTH_11[2]<0.01]
HTH_18_evalue=HTH_18[HTH_18[2]<0.01]
HTH_AraC_evalue=HTH_18[HTH_18[2]<0.01]
Lys_evalue=HTH_AraC[HTH_AraC[2]<0.01]
LytR_evalue=LytR[LytR[2]<0.01]
Response_reg_evalue=Response_reg[Response_reg[2]<0.01]
Trans_reg_evalue=Trans_reg[Trans_reg[2]<0.01]
CitT_evalue=CitT[CitT[2]<0.01]
HTH_IclR_evalue=HTH_IcIR[HTH_IcIR[2]<0.01]

HAMP_evalue=HAMP[HAMP[2]<0.01]
HATPase_c_evalue=HATPase_c[HATPase_c[2]<0.01]
HATPase_c5_evalue=HATPase_c5[HATPase_c5[2]<0.01]
HisKA_evalue=HisKA[HisKA[2]<0.01]
His_kinase_evalue=His_kinase[His_kinase[2]<0.01]
Hpt_evalue=Hpt[Hpt[2]<0.01]
PAS_evalue=PAS[PAS[2]<0.01]
HisKA_3_evalue=HisKA_3[HisKA_3[2]<0.01]
GAF_evalue=GAF[GAF[2]<0.01]
MASE1_evalue=MASE1[MASE1[2]<0.01]
CSK_evalue = CSK[CSK[2]<0.01]

```

## Concat RR and HK domains

```python
RR_domain_cat = pd.concat([CheB_evalue, GerE_evalue, HTH_11_evalue, HTH_18_evalue, HTH_AraC_evalue, Lys_evalue, LytR_evalue, Response_reg_evalue, Trans_reg_evalue, CitT_evalue, HTH_IclR_evalue])
HK_domain_cat = pd.concat([HAMP_evalue, HATPase_c_evalue, HATPase_c5_evalue, HisKA_evalue, His_kinase_evalue, Hpt_evalue, PAS_evalue, HisKA_3_evalue, GAF_evalue, MASE1_evalue, CSK_evalue])
```

## Keep unique value

```python
RR_domain_uniq = RR_domain_cat.drop_duplicates(subset=0, keep='first')
HK_domain_uniq = HK_domain_cat.drop_duplicates(subset=0, keep='first')
```

## Add field "HMM"

```python
RR_domain_uniq['hmm']="RR"
HK_domain_uniq['hmm']="HK"
```

```python
RR_HK=pd.concat([RR_domain_uniq, HK_domain_uniq])
RR_HK.to_csv("../data/e_cloacae/final_out/RR_HK_1.csv")
```

## Sort output by distance and mapped on PATRIC feature table

```python
from glob import glob
```

```python
f1 = pd.DataFrame()
f2 = pd.DataFrame()
for csv in glob('/home/akanksha/akanksha/pan-genomics/data/e_cloacae/330_features_tab/*.tab'):
    ft_i = pd.read_csv(csv, sep = '\t')
    if 'go' in ft_i.columns:
        f1 = f1.append(ft_i)
    else:
        f2 = f2.append(ft_i)
        
    
```

```python
fields = set(f1.columns) & set(f2.columns)
overall_file = f1[fields].append(f2[fields])
hmm_file = pd.read_csv('../data/e_cloacae/final_out/RR_HK_1.csv')
overall_file = hmm_file.merge(overall_file, left_on='0', right_on='patric_id', how='left')
overall_file.to_csv("../data/e_cloacae/final_out/RR_HK_merged_mapped_1.csv", sep="\t")
```

```python
overall_file = overall_file.fillna('')
```

```python
overall_file.shape
```

```python
rows = []
for gid in set(overall_file['genome_id']):
    for contig in set(overall_file.loc[(overall_file['genome_id'] == gid)]['accession']):
        tmp = overall_file.loc[(overall_file['genome_id'] == gid) & (overall_file['accession'] == contig)].sort_values(by = 'start')
        
        for i in range(1, len(tmp)):
            if tmp.iloc[i]['start'] - tmp.iloc[i-1]['end'] < 2500:
                rows.append({'genome_id':gid, 'accession':contig, 
                            'patric_id_1':tmp.iloc[i-1]['patric_id'], 'patric_id_2':tmp.iloc[i]['patric_id'],
                             'hmm_1':tmp.iloc[i-1]['hmm'], 'hmm_2':tmp.iloc[i]['hmm'],
                              'product_1':tmp.iloc[i-1]['product'], 'product_2':tmp.iloc[i]['product'],
                              'gene_1':tmp.iloc[i-1]['gene'], 'gene_2':tmp.iloc[i]['gene'],
                             'intergenic_distance':tmp.iloc[i]['start'] - tmp.iloc[i-1]['end'],
                             'start':tmp.iloc[i-1]['start'], 'end':tmp.iloc[i]['end']
                            })
    
intergenic_regions = pd.DataFrame(rows)
intergenic_regions.to_csv("../data/e_cloacae/final_out/RR_HK_output_1.csv", sep="\t")
intergenic_regions.shape
#tmp
```

```python
from collections import defaultdict
```

```python
gid2CS = defaultdict(list)

for accession in set(intergenic_regions['accession']):
    intergenic_regions_i = intergenic_regions.loc[intergenic_regions['accession'] == accession]
    gid = intergenic_regions_i['genome_id'].iloc[0]
    l1 = set()
    for gene_1 in set(intergenic_regions_i['patric_id_2']):
        if gene_1 not in l1:
            CS = intergenic_regions_i.loc[intergenic_regions_i['patric_id_1'] == gene_1].append(intergenic_regions_i.loc[intergenic_regions_i['patric_id_2'] == gene_1])
            gene_list = ['%s (%s)'%(CS.iloc[i]['patric_id_1'], CS.iloc[i]['hmm_1']) for i in range(len(CS))]
            gene_list += ['%s (%s)'%(CS.iloc[i]['patric_id_2'], CS.iloc[i]['hmm_2']) for i in range(len(CS))]
            gid2CS[gid].append(gene_list)
            l1.update(set(CS['patric_id_1'].tolist()+ CS['patric_id_2'].tolist()))

gid2CS
  
```

```python
gid2CS = []
genelist2CS = []
#f = open("RR_HK_pattern.csv","w")

for accession in set(intergenic_regions['accession']):
    intergenic_regions_i = intergenic_regions.loc[intergenic_regions['accession'] == accession]
    gid = intergenic_regions_i['genome_id'].iloc[0]
    l1 = set()
    for gene_1 in set(intergenic_regions_i['patric_id_2']):
        if gene_1 not in l1:
            CS = intergenic_regions_i.loc[intergenic_regions_i['patric_id_1'] == gene_1].append(intergenic_regions_i.loc[intergenic_regions_i['patric_id_2'] == gene_1])
            gene_list = ['%s (%s)'%(CS.iloc[i]['patric_id_1'], CS.iloc[i]['hmm_1']) for i in range(len(CS))]
            gene_list += ['%s (%s)'%(CS.iloc[i]['patric_id_2'], CS.iloc[i]['hmm_2']) for i in range(len(CS))]
            
            
            gid2CS.append(str(gid))
            genelist2CS.append(gene_list)
            l1.update(set(CS['patric_id_1'].tolist()+ CS['patric_id_2'].tolist()))
```

```python
test = pd.DataFrame({'gid': gid2CS, 'genelist': genelist2CS})
```
